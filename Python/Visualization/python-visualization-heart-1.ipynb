## Python / Visualization / Heart diseases dataset (1)

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing and cleaning dataset

```python
# Reading dataset into a pandas dataframe

cols = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num"]
df = pd.read_csv("Cleavland.csv")
df.columns = cols
df
```

```python
# Checking for missing values

df.isna().sum()
```

```python
df.info()
```

```python
# Replacing '?' with NaN

df.replace('?', np.nan, inplace = True)
df
```

```python
# Dropping rows with missing values

df.dropna(inplace = True)
df
```

```python
# Converting columns to numeric

df.apply(pd.to_numeric, errors = 'ignore')
df
```

### Data visualization

#### 1. Histogram
Objective: To visualize the distribution of continuous variables.

```python
# Plot histogram for age
df['age'].hist(bins = 20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Plot histogram for trestbps (resting blood pressure)
df['trestbps'].hist(bins = 20)
plt.title('Resting Blood Pressure Distribution')
plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.ylabel('Count')
plt.show()
```

#### 2. Dot Plots
Objective: To visualize the distribution of categorical variables.

```python
# Plot dot plot for cp (chest pain type)
sns.stripplot(x = 'cp', data = df, jitter = True)
plt.xticks(rotation = 90)
plt.title('Chest Pain Type Distribution')
plt.show()

# Plot dot plot for restecg (resting electrocardiographic results)
sns.stripplot(x = 'restecg', data = df, jitter = True)
plt.xticks(rotation = 90)
plt.title('Resting Electrocardiographic Results Distribution')
plt.show()
```

#### 3. Bar Plots
Objective: To compare the counts or mean values of a variable across different categories.

```python
# Plot dot plot for cp (chest pain type)
sns.stripplot(x = 'cp', data = df, jitter = True)
plt.xticks(rotation = 90)
plt.title('Chest Pain Type Distribution')
plt.show()

# Plot dot plot for restecg (resting electrocardiographic results)
sns.stripplot(x = 'restecg', data = df, jitter = True)
plt.xticks(rotation = 90)
plt.title('Resting Electrocardiographic Results Distribution')
plt.show()
```

#### 4. Line Charts
Objective: To visualize trends or changes over time or another continuous variable.

```python
# Plot line chart for mean thalach (maximum heart rate achieved) by age
mean_thalach = df.groupby('age')['thalach'].mean()
mean_thalach.plot(kind = 'line')
plt.title('Mean Maximum Heart Rate by Age')
plt.xlabel('Age')
plt.ylabel('Mean Maximum Heart Rate')
plt.show()
```

#### 5. Add Histogram and Scatter plot to Box plot
Objective: To visualize the distribution, outliers, and density of a continuous variable.

```python
# Box plot with histogram and scatter plot
plt.figure(figsize = (10, 6))
plt.subplot(2, 2, 1)
sns.boxplot(data = df['age'])
plt.title('Age')

plt.subplot(2, 2, 2)
sns.histplot(data = df, x = 'age', kde = True)
plt.title('Age Distribution')

plt.subplot(2, 2, 3)
sns.scatterplot(data = df, x = 'age', y = 'thalach')
plt.title('Age vs Maximum Heart Rate')

plt.tight_layout()
plt.show()
```
