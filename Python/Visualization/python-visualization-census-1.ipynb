## Python / Visualization / Census bureau dataset (1)

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing and cleaning dataset

```python
# Reading dataset into a pandas dataframe

col = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'target']
df = pd.read_csv('adult.csv')
df.columns = col
df
```

```python
# Checking for missing values

df.isna().sum()
```

```python
# Replacing '?' with NaN values and dropping the rows with NaN values

df.replace('?', np.nan, inplace = True)
df.dropna(inplace = True)
```

```python
# Converting columns to numeric

df.apply(pd.to_numeric, errors = 'ignore')
df
```

### Data visualization

#### 1. Histogram
Objective: To visualize the distribution of continuous variables.

```python
# Plot histogram for age
df['age'].hist(bins = 20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Plot histogram for hours-per-week
df['hours-per-week'].hist(bins = 20)
plt.title('Hours per Week Distribution')
plt.xlabel('Hours per Week')
plt.ylabel('Count')
plt.show()
```

#### 2. Dot Plots
Objective: To visualize the distribution of categorical variables.

```python
# Plot dot plot for education
sns.stripplot(x = 'education', data = df, jitter = True)
plt.xticks(rotation = 90)
plt.title('Education Distribution')
plt.show()

# Plot dot plot for occupation
sns.stripplot(x = 'occupation', data = df, jitter = True)
plt.xticks(rotation = 90)
plt.title('Occupation Distribution')
plt.show()
```

#### 3. Bar Plots
Objective: To compare the counts or mean values of a variable across different categories.

```python
# Plot bar plot for race
race_counts = df['race'].value_counts()
race_counts.plot.bar()
plt.title('Race Distribution')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()

# Plot bar plot for mean age by marital status
mean_age = df.groupby('marital-status')['age'].mean()
mean_age.plot.bar()
plt.title('Mean Age by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Mean Age')
plt.show()
```

#### 4. Line Charts
Objective: To visualize trends or changes over time or another continuous variable.

```python
# Plot line chart for mean education-num by age
mean_education_num = df.groupby('age')['education-num'].mean()
mean_education_num.plot(kind = 'line')
plt.title('Mean Education Level by Age')
plt.xlabel('Age')
plt.ylabel('Mean Education Number')
plt.show()
```

#### 5. Add Histogram and Scatter plot to Box plot
Objective: To visualize the distribution, outliers, and density of a continuous variable.

```python
# Box plot with histogram and scatter plot
plt.figure(figsize = (10, 6))
plt.subplot(2, 2, 1)
sns.boxplot(data = df['age'])
plt.title('Age')

plt.subplot(2, 2, 2)
sns.histplot(data = df, x = 'age', kde = True)
plt.title('Age Distribution')

plt.subplot(2, 2, 3)
sns.scatterplot(data = df, x = 'age', y = 'hours-per-week')
plt.title('Age vs Hours per Week')

plt.tight_layout()
plt.show()
```
