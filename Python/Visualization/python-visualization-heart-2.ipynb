## Python / Visualization / Heart diseases dataset (2)

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing and cleaning dataset

```python
# Reading dataset into a pandas dataframe

cols = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num"]
df = pd.read_csv("Cleavland.csv")
df.columns = cols
df
```

```python
# Checking for missing values

df.isna().sum()
```

```python
df.info()
```

```python
# Replacing '?' with NaN

df.replace('?', np.nan, inplace = True)
df
```

```python
# Dropping rows with missing values

df.dropna(inplace = True)
df
```

```python
# Converting columns to numeric

df.apply(pd.to_numeric, errors = 'ignore')
df
```

### Data visualization

#### 1. Histogram
Objective: To visualize the distribution of continuous variables.

```python
# Plot histogram for age
df['age'].hist(bins = 20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Plot histogram for trestbps (resting blood pressure)
df['trestbps'].hist(bins = 20)
plt.title('Resting Blood Pressure Distribution')
plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.ylabel('Count')
plt.show()
```

#### 2. Pie Charts
Objective: To visualize the proportions of different categories in a categorical variable.

```python
# Plot pie chart for sex
sex_counts = df['sex'].value_counts()
plt.pie(sex_counts, labels = ['Female', 'Male'], autopct = '%1.1f%%')
plt.title('Sex Distribution')
plt.axis('equal')  # To make the pie chart circular
plt.show()

# Get the unique categories in the 'num' column
diagnosis_categories = df['num'].unique()

# Plot pie chart for diagnosis
diagnosis_counts = df['num'].value_counts()
plt.pie(diagnosis_counts, labels = diagnosis_categories, autopct = '%1.1f%%')
plt.title('Diagnosis Distribution')
plt.axis('equal')
plt.show()
```

#### 3. Box Plots
Objective: To visualize the distribution of a continuous variable, including median, quartiles, and outliers.

```python
# Plot box plot for age
plt.figure(figsize = (8, 6))
sns.boxplot(data = df['age'])
plt.title('Age Distribution')
plt.show()

# Plot box plot for trestbps (resting blood pressure)
plt.figure(figsize = (8, 6))
sns.boxplot(data = df['trestbps'])
plt.title('Resting Blood Pressure Distribution')
plt.show()
```

#### 4. Scatter Plots
Objective: To visualize the relationship between two continuous variables.

```python
# Plot scatter plot for age vs. thalach (maximum heart rate achieved)
plt.figure(figsize = (8, 6))
plt.scatter(df['age'], df['thalach'])
plt.title('Age vs. Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.show()

# Plot scatter plot for trestbps (resting blood pressure) vs. chol (serum cholestoral)
plt.figure(figsize = (8, 6))
plt.scatter(df['trestbps'], df['chol'])
plt.title('Resting Blood Pressure vs. Serum Cholestoral')
plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.ylabel('Serum Cholestoral (mg/dl)')
plt.show()
```

#### 5. Add Boxplots to a Scatterplot
Objective: To visualize the distribution of a continuous variable and its relationship with another continuous variable.

```python
# Add boxplots to a scatterplot
plt.figure(figsize = (10, 6))
sns.scatterplot(data = df, x = 'age', y = 'thalach')
sns.boxplot(data = df['thalach'], orient = 'v')
plt.title('Age vs. Maximum Heart Rate with Boxplots')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.show()
```
