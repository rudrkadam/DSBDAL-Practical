## Python / Models / Iris dataset

### Importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

### Importing and cleaning dataset

```python
# Reading dataset into a pandas dataframe

df = pd.read_csv("Iris.csv")
df
```

```python
# Redefining column names

cols = ["sepal length in cm", "sepal width in cm", "petal length in cm", "petal width in cm", "class"]
df.columns = cols
df
```

```python
# Checking for missing values

df.isna().sum()
```

```python
# Replacing '?' with NaN and dropping rows with NaN values

df.replace('?', np.nan, inplace = True)
df.dropna(inplace = True)
```

```python
# Converting columns to numeric

df.apply(pd.to_numeric, errors = 'ignore')
df
```

```python
numerical_cols = df.select_dtypes(include = np.number).columns
```

### Data analysis

```python
# Creating two variables for analysis

x = df.drop('class', axis = 1)
y = df['class']
```

```python
# Scaling data

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
# Creating train and test sets

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 42)
```

```python
# Logistic Regression

model1 = LogisticRegression()
model1.fit(x_train, y_train)
model1.predict(x_test)
print(accuracy_score(y_test, model1.predict(x_test)))
```

```python
# Naive Bayes

model2 = GaussianNB()
model2.fit(x_train, y_train)
model2.predict(x_test)
print(accuracy_score(y_test, model2.predict(x_test)))
```
