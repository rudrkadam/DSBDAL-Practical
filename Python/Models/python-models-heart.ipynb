## Python / Models / Heart diseases dataset

### Importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

### Importing and cleaning dataset

```python
# Reading dataset into a pandas dataframe

df = pd.read_csv("Cleavland.csv")
df
```

```python
# Redefining column names

cols = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num"]
df.columns = cols
df
```

```python
# Checking for missing values

df.isna().sum()
```

```python
# Replacing '?' with NaN

df.replace('?', np.nan, inplace = True)
df
```

```python
# Dropping rows with missing values

df.dropna(inplace = True)
df
```

```python
# Converting columns to numeric

df.apply(pd.to_numeric, errors = 'ignore')
df
```

### Data analysis

```python
# Creating two variables for analysis

x = df.drop('num', axis = 1)
y = df['num']
```

```python
# Scaling data

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
# Creating train and test sets

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 42)
```

```python
# Logistic Regression

model1 = LogisticRegression()
model1.fit(x_train, y_train)
model1.predict(x_test)
print(accuracy_score(y_test, model1.predict(x_test)))
```

```python
# K-Nearest Neighbors

model2 = KNeighborsClassifier()
model2.fit(x_train, y_train)
model2.predict(x_test)
print(accuracy_score(y_test, model2.predict(x_test)))
```
