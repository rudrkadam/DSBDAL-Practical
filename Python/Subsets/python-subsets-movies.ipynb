## Python / Subsets / Movies dataset

### Importing libraries

```python
import pandas as pd
```

```python
# Reading dataset into a pandas dataframe

DataFrame = pd.read_csv('movies_metadata.csv', low_memory = False)
DataFrame
```

### Creating a subset for different languages

```python
# Subset for original language being English

Subset1 = DataFrame[DataFrame['original_language'] == 'en']
Subset1
```

```python
# Subset for original language being Italian

Subset2 = DataFrame[DataFrame['original_language'] == 'it']
Subset2
```

```python
# Subset for original language being Dutch

Subset3 = DataFrame[DataFrame['original_language'] == 'nl']
Subset3
```

### Merging subsets

```python
Merged1 = pd.concat([Subset1, Subset2])
Merged1
```

```python
Merged2 = pd.concat([Subset2,Subset3])
Merged2
```

```python
Merged3 = pd.concat([Subset1, Subset2, Subset3])
Merged3
```

### Sorting data using customer rating

```python
sorted_data = DataFrame.sort_values(by = ['vote_average'], ascending = False)
sorted_data
```

### Transposing data

```python
transpose_data = DataFrame.transpose()
transpose_data
```

### Melting data to long format

```python
MeltedDf = pd.melt(DataFrame, id_vars = ['id','original_language'], var_name = 'Metric', value_name = 'value')
MeltedDf
```

### Casting data to wide format

```python
# Select the columns you want to include in the wide format
cols_to_keep = ['adult', 'budget', 'genres', 'original_title', 'revenue', 'runtime', 'vote_average', 'vote_count']

# Reset index to create column named 'index'
df = DataFrame.reset_index()  

# Melt the DataFrame to long format
df_long = pd.melt(df[cols_to_keep+['index']], id_vars='index', value_vars=cols_to_keep, var_name='column_name', value_name='value')

# Pivot the long format DataFrame to wide format
df_wide = df_long.pivot(index='index', columns='column_name', values='value')

df_wide
```
