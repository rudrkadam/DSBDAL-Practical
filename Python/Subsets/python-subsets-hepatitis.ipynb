## Python / Subsets / Hepatitis dataset

### Importing libraries

```python
import pandas as pd
```

```python
# Reading dataset into a pandas dataframe

col = ['Class', 'AGE', 'SEX', 'STEROID', 'ANTIVIRALS', 'FATIGUE', 'MALAISE', 'ANOREXIA', 'LIVER BIG', 'LIVER FIRM', 'SPLEEN PALPABLE', 'SPIDERS', 'ASCITES', 'VARICES', 'BILIRUBIN', 'ALK PHOSPHATE', 'SGOT', 'ALBUMIN', 'PROTIME', 'HISTOLOGY']
DataFrame = pd.read_csv('hepatitis.csv')
DataFrame.columns = col
DataFrame
```

### Creating subsets for different sex

```python
# Unique sex values in the dataset

x = pd.unique(DataFrame['SEX'])
x
```

```python
# Subset for SEX 1

SEX1 = DataFrame[DataFrame['SEX'] == 1]
SEX1
```

```python
# Subset for SEX 2

SEX2 = DataFrame[DataFrame['SEX'] == 2]
SEX2
```

### Merging subsets

```python
Merged1 = pd.concat([SEX1, SEX2])
Merged1
```

### Sorting data using age, SGOT, PROTIME

```python
Sorted_Age = DataFrame.sort_values('AGE')
Sorted_Age
```

```python
Sorted_SGOT = DataFrame.sort_values('SGOT')
Sorted_SGOT
```

```python
Sorted_PROTIME = DataFrame.sort_values('PROTIME')
Sorted_PROTIME
```

### Transposing data

```python
transpose_data = DataFrame.transpose()
transpose_data
```

### Melting data to long format

```python
MeltedDf = pd.melt(DataFrame, id_vars = ['Class', 'AGE'], value_name = 'value', var_name = 'Metrics')
MeltedDf
```

### Casting data to wide format

```python
CastedDf = pd.pivot_table(DataFrame, index=['Class', 'AGE'], values = 'HISTOLOGY')
CastedDf
```
