## Python / Subsets / Facebook dataset

### Importing libraries

```python
import pandas as pd
```

### Information about data   

```python
# Reading dataset into a pandas dataframe

dataframe = pd.read_csv('dataset_Facebook.csv', sep=';')
dataframe.head()
```

```python
# Information about dataset

dataframe.info()
```

```python
# Description of dataframe

dataframe.describe()
```

```python
# Columns and their data types

print('shape of data - ', dataframe.shape)
print('Columns - ', dataframe.columns ) 
print('Data Types - ', dataframe.dtypes)
```

### Counting and removing null values

```python
# Total number of null values 

if dataframe[dataframe == 0].count().sum() == 0:
    print("No Null values present!")
else:
    print("Total Null values are ", dataframe[dataframe == 0].count().sum()) 
```

```python
dataframe.dropna(inplace=True)
# mean = dataframe.abs.mean()
# dataframe.fillna(mean)
```

```python
dataframe.isnull().sum().sum()
```

### Creating a subset for type of post

```python
# Subset for type 'Photo'

subset_1 = dataframe.loc[dataframe['Type'] == 'Photo'] 
subset_1 
```

```python
# Subset for type 'Video'

subset_2 = dataframe.loc[dataframe['Type'] == 'Video']
subset_2
```

### Merging two subsets 

```python
pd.merge(subset_1, subset_2, on = 'like', how = 'inner')
```

### Sorting data on total likes

```python
sorted_df = dataframe.sort_values(by = ["like"], ascending = True)
sorted_df
```

### Transposing data

```python
transposed_df = dataframe.transpose()
transposed_df
```

### Melting data to long format

```python
melted_df = pd.melt(frame = dataframe, id_vars = "Category", value_vars = ["like"])
melted_df
```

### Casting data to wide format

```python
casted_df = dataframe.pivot_table(index = ['Category'], columns = "Paid", values = "like")
casted_df
```
