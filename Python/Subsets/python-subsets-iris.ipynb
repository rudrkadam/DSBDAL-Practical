## Python / Subsets / Iris dataset

### Importing libraries

```python
import pandas as pd
```

```python
# Reading dataset into a pandas dataframe

df = pd.read_csv('Iris.csv')
df
```

```python
# Columns in the dataset

df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'species']
print(df)
```

```python
# Description of dataframe

df.describe()
```

### Creating a subset for different species

```python
# Unique species in the dataset

x = pd.unique(df['species'])
x
```

```python
# Subset for specoes 'Iris-virginica'

subset1 = df.loc[df['species'] == 'Iris-virginica']
subset1
```

```python
# Subset for species 'Iris-versicolor'

subset2 = df.loc[df['species'] == 'Iris-versicolor']
subset2
```

```python
# Subset for species 'Iris-setosa'

subset3 = df.loc[df['species'] == 'Iris-setosa']
subset3
```

### Merging subsets

```python
merge_ds = pd.merge(subset1, subset2, on = 'sepal length', how = 'inner')
merge_ds
```

```python
merge_ds2 = pd.merge(subset2, subset3, on = 'sepal width', how = 'inner')
merge_ds2
```

### Sorting data on petal length

```python
sorted_data = df.sort_values(by = ['petal length'], ascending = True)
sorted_data
```

### Transposing data

```python
transpose_data = df.transpose()
transpose_data
```

### Melting data to long format

```python
melted_df = pd.melt(frame = df, id_vars=['sepal length','petal length','species'])
melted_df
```

### Casting data to wide format

```python
casted_df = df.pivot_table(df, index = ['species'])
casted_df
```
